### Running SXDM

#### Set Up
1. Run the `SXMD.py` file in jupyter notebook
1. Name a variable (preferably your sample name and run the SXDM class on it.
    * **Example:** `sample_1=SXDM(fov=1,scanlist=[1,2,3,4],scanangles=[15.1,15.4,13.2,20.3],sample='sample_1',folder_path='/home/will/Desktop/test/')`
    * Acceptable values for all these variables are listed in the SXDM() description.
    
1. If you have already run this command once errors will pop up telling you something similar to 

`Files will be rewritten. Will not continue. File System: 
/home/will/Desktop/Cycled/WIP^/1-Sample/1-LocationMat/FOV1/`

* the program auto formats the work you have just done and instead of rewriting overold files the program won't continue just incase you hit this command by mistake.
    
#### Summed Diffraction
1. For your first pass run: `sample_1.summeddif_cal('n','n')`
   * The first 'n' is asking "Are you performing the summed diffraction after your ROI Calculation?". Since you are not this value should be set to 'n'. 
   * The second 'n' is asking "Have you already run summeddif_cal once and just want to change the contrast?" Since we have not run this script yet this value should be set to 'n'
1. Sometimes the program doesn't get the contrast right on the first try. If you want to change the contrast just set the second variable equal to 'y' and add a variable named vmin or vmax or both. 
    * **Example:** `sample_1.summeddif_cal('n','y',vmax=1200)`
    
##### Summed Diffraction Does't Look Right:
1. Set the muliplier to a different values. Setting it to equal 2 usually works. 
1. The user can also set the background_sub_reverse variable to equal 'y'
    
#### ROI Viewer: Pre

1. Once the `sample_1.summeddif_cal` script is done it will output the dimensions of your ROI
1. Use these dimensions for the roi_crop_pre script
1. Run `sample_1.roi_crop_pre(0,Max_x,0,Max_y)`  
    * **Example:** `sample_1.roi_crop_pre(0,27,0,34)`
1. Once complete you may change all 4 values for a sutable ROI crop
    * **Example:** `sample_1.roi_crop_pre(15,20,3,29)`
1. To recalculate the summed diffraction on the new ROI you created rerun the `sample_1.summeddif_cal()` script, but changing the first varable to equal 'y'.
    * **Example:** `sample_1.summeddif_calc('y','n',vmax=1200)`
    
#### ROI Viewer: Post

This script adds background noise to the ROI scanning angles that needed to be moved inorder for alignment. It is only nessessary to run if you want an accurate result for the `sample_1.gaussiancheck()` scritp.

1. Run `sample_1.roi_crop_post(val1,val2,val3,show_new_summeddif='y',vmax=int)` 
    * Read desctiption in script on what each value represents

#### Creating Your Own ROI: 
1. Once the `sample_1.summeddif_cal` is done you can run the `sample_1.new_roi_step1(tup_array)` to define new ROI regions
1. the maximum is 4 regions. The first 3 are for and RGB image and the last is for a total ROI
1. Step 1 allows you to determine the ROI's for the calculation while `sample_1.new_roi_step2()` will create roi variables of your defined regions. 

#### It Is Strongly Advised That You Create Your Own Master ROI
Since the nanoprobe doesn't handle hot/dead pixles well you can run the `sample_1.new_roi_step1()` and `sample_1.new_roi_step2()` without any input arguments to create a `sample_1.roi_master` which is a ROI map for the entire detector subtracting the hot/dead pixles. 

#### Gaussian Check

1. Run `sample_1.gaussiancheck()`
   * This will show you if your scans have a gaussian distribution of intensities. 


#### Determining Chi Limits

1. Run `sample_1.chi_step1(scan_number_str)`
    * If needed add a vmin or vmax variable to this script in order for you to see the signal coming from these scans.
1. Run `sample_1.chi_step2(Index_of_First_Image_You_Want_To_Use,Index_of_Last_Image_You_Want_To_Use)`
    * You can change the vmin and vmax values to match the ones in chi_step1
    * Change the first_vert and second_vert values to where they are hitting the same spot in the data in both images.
1. Run `Chi=sample_1.chi_step3(Size_of_Pixles_in_um, Total_Angle_in_Degrees_of_Images_From_Step_2)`


#### Chi and 2Theta Maps: Must Complete The sample_1.roi_crop_pre() Function Before Running

1. Run `sample_1.Chi_TTheta_Maps(detector_angle,Chi,sub_fov=1,save='y')`
    * detector_angle is equal to the starting angle of the detector
    * Chi can be left as a variable since we assigned it in `sample_1.chi_step3()`
    * sub_fov is equal to the field of view inside each field of view you are looking at. If you crop a particular field of view you can set the sub_fov to equal 2 allowing the program to not overwrite your previous data performed on a full FOV
    * if the images do not turn out right you will have to look at the saved images and determine the best `summed_ims_vmax` and `stdev_min` values. 
    * Look at the documentation inside this function for more information about all its variables


#### Determing The Instrumental Broadening In Pixles 

1. Run `sample_1.beam_size_in_pixles(D_um,d_rN_nm,Starting_Angle_of_Detector,Kev)` 
    * the D_um is the diameter of the zone plate in microns **Example: D_um = 20**
    * the d_rN_nm is the d spacing of the outer most zone plate **Example: d_rN_nm = 20**
    * the Starting_Angle_of_Detector is exactly what you would think it is **Example: 16.7**
    * Kev is the electron volts used in the experiment **Example: 9**
