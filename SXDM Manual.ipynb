### File System Setup

1. Use Jupyter Notebook or Jupyter Lab to run `%run .../SXDM.py`  - It will load the SXDM class as well as all the functions used for analysis

1. Run `SXDM_folder_management(folder_path=,sample_number=)`. This function allows the User to specify were they would like their file system to be set up. The User has to define two variables. The first being the folder_locaion and second being the sample_number. Use the shift + tab short key to learn more about this function.
    * **Example:** `SXDM_folder_management(folder_path='/home/will/Desktop/test/',sample_number=3)`
    * folder_path is the path where you want the folder system to be created.
    * sample_number is how many samples you have. For a battery you may have a Pristine, Charged, and Discharged sample. If this is the case then sample_number=3.
    * Any folder with an * at the end of it is acceptable for you to change without any problems. Although it would be much easier to run through the program once without changing any file names.
    * If you do change and of the `x-Sample` folder names you will need to edit the `SXDM.py` file.

1. Inside the file system, which you just created open up the `Beam_Line_Data*` folder. 
    * Inside you should see a folder called `SXDM`. Open it
    * There should now be a folder called `APS_Filename*`. Open it or if you wish you can change the name first then open it. 
    * Open this newly named folder. There should be a folders called `1-Sample`,`2-Sample`,... and so on depending on what value you gave `sample_number` in addition to a folder called `zDetector`. 
    * Open ```1-Sample```. Inside you should see a folder called `MDA`.
    * Go to your data the APS gave you. Go to your version of sample 1 Images and make a link to this folder.
    * Place this linked folder into the `1-Sample` folder, next to the `MDA` folder.
    * Now you can go back a directory and open the `MDA` folder for the sample you just chose. This is where you can copy and paste all .mda files used during the experiment.
    * Repeat the Linked Images folder movement and MDA folder copying until all samples are complete. 
1. Inside the file system created open the `zDetector` folder.
    * This should be one inside the `APS_Filename*` folder. Place the Beamline scan folder which corresponds to the scan in which you determined the pixle size of the CCD Detector. Keep this scan with it's original scan name.
    
##### Congrats! You have finished the initial file system set up!

### Running SXDM

#### Set Up
1. Run the `SXMD.py` file in jupyter notebook
1. Name a variable (preferably your sample name and run the SXDM class on it.
    * **Example:** `sample_1=SXDM(fov=1,scanlist=['1','2','3','4'],scanangles=[15.1,15.4,13.2,20.3],sample='sample_1',folder_path='/home/will/Desktop/test/')`
    * Acceptable values for all these variables are listed in the SXDM() description.
    
#### Summed Diffraction
1. For your first pass run: `sample_1.summeddif_cal('n','n')`
   * The first 'n' is asking "Are you performing the summed diffraction after your ROI Calculation?". Since you are not this value should be set to 'n'. 
   * The second 'n' is asking "Have you already run summeddif_cal once and just want to change the contrast?" Since we have not run this script yet this value should be set to 'n'
1. Sometimes the program doesn't get the contrast right on the first try. If you want to change the contrast just set the second variable equal to 'y' and add a variable named vmin or vmax or both. 
    * **Example:** `sample_1.summeddif_cal('n','y',vmax=1200)`
    
#### ROI Viewer: Pre

1. Once the `self.summeddif_cal` script is done it will output the dimensions of your ROI
1. Use these dimensions for the roi_crop_pre script
1. Run `sample_1.roi_crop_pre(0,Max_x,0,Max_y)`  
    * **Example:** `sample_1.roi_crop_pre(0,27,0,34)`
1. Once complete you may change the all 4 values for a sutable ROI crop
    * **Example:** `sample_1.roi_crop_pre(15,20,3,29)`
    
#### ROI Viewer: Post

This script adds background noise to the ROI scanning angles that needed to be moved inorder for alignment. It is only nessessary to run if you want an accurate result for the `.gaussiancheck()` scritp.

1. Run `sample_1.roi_crop_post(val1,val2,val3)` 
    * Read desctiption in script on what each value represents
    
#### Gaussian Check

1. Run `sample_1.gaussiancheck()`


#### Determining Chi Limits

1. Run `sample_1.chi_step1(scan_number_str)`
    * If needed add a vmin or vmax variable to this script in order for you to see the signal coming from these scans.
1. Run `sample_1.chi_step2(Index_of_First_Image_You_Want_To_Use,Index_of_Last_Image_You_Want_To_Use)`
    * You can change the vmin and vmax values to match the ones in chi_step1
    * Change the first_vert and second_vert values to where they are hitting the same spot in the data in both images.
1. Run `Chi=sample_1.chi_step3(Size_of_Pixles_in_um, Total_Angle_in_Degrees_of_Images_From_Step_2)`


#### Chi and 2Theta Maps

1. Run `sample_1.Chi_TTheta_Maps(detector_angle,Chi,sub_fov=1,save='y')`
    * detector_angle is equal to the starting angle of the detector
    * Chi can be left as a variable since we assigned it in `sample_1.chi_step3()`
    * sub_fov is equal to the field of view inside each field of view you are looking at. If you crop a particular field of view you can set the sub_fov to equal 2 allowing the program to not overwrite your previous data performed on a full FOV
    * if the images do not turn out right you will have to look at the saved images and determine the best `summed_ims_vmax` and `stdev_min` values. 
    * Look at the documentation inside this function for more information about all its variables
